```python
import math
import matplotlib.pyplot as plt
import numpy as np
import random
X = [1, 1]
f = lambda X: X[0]**2 + X[1]**2
df = lambda X: [2*X[0], 2*X[1]]

resolucion = 30
rango=2.5
X=np.linspace(-rango, rango,resolucion)
Y=np.linspace(-rango, rango,resolucion)
Z=np.zeros((resolucion,resolucion))
for ix, x in enumerate(X):
  for iy, y in enumerate(Y):
    Z[ix,iy] = f([x,y])

contour = plt.contour(X,Y,Z, resolucion)
cbar=plt.colorbar(contour)

# Generar punto de partida aleatorio
P=[random.uniform(-rango,rango), random.uniform(-rango, rango)]
plt.plot(P[0], P[1], "o", c="green")
TA=0.05

for _ in range(500):
  grad=df(P)
  P[0], P[1] = P[0] - TA*grad[0], P[1] - TA*grad[1]
  plt.plot(P[0], P[1], "o", c="red")

plt.plot(P[0], P[1], "o", c="white")
print("Solución:", P, f(P) )
```

se puede poner una tasa de aprendizaje que se ajuste con el gradiente

en valor absoluto

también se puede  añadir un término de "momento"


```python
def gradient_descent(start, step = 0.001, tresh = 0.01, max_steps = 1e5):
  min = start
  i = 0
  while(True):
    if i > max_steps:
      return min

    new_point = np.array(min) - step * np.array(df(min))
    if (new_point - min).all() < tresh:
      return min

    min = new_point
    i+=1


min = gradient_descent([1,1])
print(f(min))
```
